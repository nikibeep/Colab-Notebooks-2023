1. Write a function “MyPassion()”which takes in a sentence (a minimum of 5 words)
and returns back a dictionary indicating the “mode” of each word. &quot;mode&quot; of a set of
data is the data point that appears most frequently , mode of &quot; Taradiddle&quot; is &quot;d&quot;.
Set of data can, in fact, have multiple modes, so &quot;tomato” has “t”, ”o” as mode. Note
that if all data appears the same number of times there is no mode, like in “dog”.
Also create and display a set that has modes and its frequencies

```python
def MyPassion(sen):
    mode_dict = {}
    word_cnt = {}
    max_val = ''
    freq_list = []
    all_value_same = False
    word_list = sen.split(' ')
    
    for word in word_list:
        max_val = ''
        word_cnt={}
        for ch in word:
            if ch not in word_cnt.keys():
                word_cnt[ch] = 1
            else:
                word_cnt[ch]+=1
                
        max_val = max(word_cnt.values())
        all_value_sam = all(val == max_val for val in word_cnt.values())
        freq_list = [k for k,v in word_cnt.items() if v == max_val]

        if not all_value_sam:
            mode_dict[word] = freq_list
        else:
            mode_dict[word] = []

    return mode_dict

def input_value():
    sentence = input("Enter the sentence: >> ")
    high_fq = MyPassion(sentence)
    print(high_fq)

input_value()
      



```
