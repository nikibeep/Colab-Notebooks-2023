1. Write a python program to read a list of numbers.
 Find the mean and median and print.
 Remove all duplicates present in the list.
 Remove the numbers that occur more than 3 times

```python
list1=[]
n= int(input("enter the number of elements"))
for i in range (0,n):
  e=int(input())
  list1.append(e)
  l=len(list1)
sum=0
for i in list1:
  sum += i
list1.sort()    #median should be taken after sorting. 
print("sorted list:", list1)
if(l%2==0):
  m1=int(l/2)  #m1=index
  print("median: ",(list1[m1-1]+list1[m1])/2)
else:
  m1=int(n/2)
  print("Median", list1[m1])
print("mean is", sum/n)
 
```

Remove all duplicates present in the list

```python
l=[]
l2=[]
l3=[]
n= int(input("enter the number of elements"))
for i in range (0,n):
  e=int(input())
  l.append(e)
print("Original List: ", l)
for i in l:
  if i not in l2:
    l2.append(i)
  if l.count(i)<3:
    l3.append(i)
print("List after removing duplicate elements: ", l2 )
print("after removing the elements count greaterthan 3:", l3)
```

2. Write a python program to find the reverse pair of words in the given word list.

```python
l=[]
string=input("Enter your words list :").split()
print("\n The reverse pair of words if any ")

for x in string:
  if x[::-1] in string and x not in l: #x[::-1] gives reverse number and l list is used to print unique values
    print("\n",x,"and",x[::-1])
    l.append(x)
    l.append(x[::-1])

```

3. Get a text from the user. Add “ed” at the of the string. Length of the string must be atleast 5.
If the string already has “ed” at the end, then add ing. If the length of the string is less than
6, print the string unchanged with a message.

```python
inp=input("Enter Your Word Here: ")
length=len(inp)
if length <5:
  print(inp)
  print("your string is too small")
  
elif length>=5 and inp[-2:]=="ed":
  inp1=inp[:length - 2]
  inp=inp1+"ing"
  print(inp)

elif length <6 and inp[-2:]!="ed":
  print("String not changed:", inp)
```

4. Get a string from the user. Encrypt the string and print both the strings. Use the following
rules to encrypt the string
 Add a number to each alphabet in the given string to generate its corresponding
letter
 Number should be taken from the user

```python
str1=input("Enter the String: ")
num=int(input("Enter the Code: "))
str2=str1.upper()
for i in str2:
  if((ord(i)+num)>90):
    n=(ord(i)+num)-(26)
  else:
    n=ord(i)+num
  print(chr(n),end="")
```
