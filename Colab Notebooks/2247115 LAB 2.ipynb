1. Given the list of CGPA of students CGPA=[8,7,9,5,7,6,9,8,6,6] Write: 
a) An expression that evaluates to the number of 9 CGPA 
b) A statement that changes the last CGPA to 10
c) An expression that evaluates to the maximum and minimum CGPA 
d) A statement that sorts the list CGPA 
e) An expression that evaluates to the average CGPA

```python
cgpa=[8,7,9,5,7,6,9,8,6,6] 
print("9 present in the list:", cgpa.count(9), "times")
cgpa[-1]=10    #last element change to 10
print("CGPA after adding 10:", cgpa)
print("Maximum CGPA:", max(cgpa))    # maximum value in the list
print("Minimum CGPA:", min(cgpa))    # minum value in the list
cgpa.sort()
print("Sorted list:", cgpa)
print("Average CGPA:", sum(cgpa)/len(cgpa))  # average of the list
```
