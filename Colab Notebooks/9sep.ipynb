```python
#structured arrays
name=['Aryan','Kevin','Alvis','Adithya']
age=[20,21,22,21]
weight=[48.0,51.0,57.0,63.0]
```

```python
x=np.zeros(4,dtype=int)
x
```

```python
data=np.zeros(4,dtype={'names':('name','age','weight'),'formats':('U10','i4','f8')})
```

```python
data['name']=name
data['age']=age
data['weight']=weight
print(data)
```

```python
#refering values by index or by name
print(data['name'])#getting all names
print(data[0])#first row of data
```

```python
#referring values by index or by name
print(data['name'])#getting all names
print(data[0]) #first row of adata
print(data[-1]['name']) #getting the name from the last row
```

```python
data[data['age']<30]['name']
```

```python
#crating Structured arrays
np.dtype({'names':('name','age','weight'),'formats':('U10','i4','f8')})
```

```python
np.dtype({'names':('name','age','weight'),'formats':((np.str_,10),int,np.float32)})
```

```python
np.dtype([('name','S10'),('age','i4'),('weight','f8')])
```
