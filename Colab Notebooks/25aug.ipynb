```python
import csv
courses=[('OS',1,'Sep'),('DL',2,'Sep'),('Stat',3,'Sep')]
cf=open('student.csv','w',newline='')#write Mode
obj=csv.writer(cf)
for courses in courses:
  obj.writerow(courses) #writerow
cf.close
```

```python
import csv
courses=[('PY',4,'Sep'),('C',5,'Sep')]
cf=open('student.csv','r',newline='')
obj=csv.writer(cf)
for courses in courses:
    obj.writerows(courses)
cf.close()
```

```python
cf=open('student.csv','r',newline='')#read
obj=csv.reader(cf)
for row in obj:
  print(row)
```
