```python
import re
def loveforicecreams(s):
  num = len(re.findall(r'\bbutterscotch\b', s))
  return num
print(loveforicecreams('butterscotch in butterscotch at butterscotch parlour'))
```

```python
import re
from functools import reduce
def Linkconsonants(a):
  b = re.findall(" ([b, c, d, f, g, h, j, k, l, m, n, p, q, r, s, t, v, w, x, y, z, ',', '.'].+?[b, c, d, f, g, h, j, k, l, m, n, p, q, r, s, t, v, w, x, y, z, ',', '.']) ", a.lower())
  ending_in_vowel = r'(\b\w+[AaEeIiOoUu]\b)?'
  begin_in_vowel = r'(\b[AaEeIiOoUu]\w+\b)?'
  ignore = [b for b in re.findall(begin_in_vowel, a) if b]
  ignore.extend([b for b in re.findall(ending_in_vowel, a) if b])
  result = [word for word in a.split() if word not in ignore]
  fin=set(a.split()).difference(result)
  nl=list(fin)
  for z in result:
    print(z,end=" {}".format("--> consonants end and start \n"))
  for ele in fin:
    print(ele,end=" {}".format("aint consonants end and start(false)\n"))
  count=reduce(lambda x,y:x+1,nl,0)
  if(count>=1):
    for x in result:
      print("(true) {} ends with and starts with consonants".format(x))
print(Linkconsonants("These are crazy"))
```
