```python
import statistics
def Remove(duplicate):
    final_list = []
    for num in duplicate:
        if num not in final_list:
            final_list.append(num)
    return final_list
    
duplicate = [2, 4, 10, 20, 5, 2, 20, 4]
print(Remove(duplicate))
z = (Remove(duplicate))
x = statistics.mean(z)
y = statistics.median([2, 4, 10, 20, 5, 2, 20, 4])
print("Mean is :", x)
print("Median is :",y)
res = [i for i in duplicate if duplicate.count(i) == 1]
print(res)
```
